```python
import csv
from pathlib import Path

src = "../data/Produkt_export_Aug_2025__fixed.csv"   # use your fixed file from before
cleaned = "../data/Produkt_export_Aug_2025__repaired.csv"

with open(src, newline="", encoding="utf-8-sig") as fin, \
     open(cleaned, "w", newline="", encoding="utf-8") as fout:
    r = csv.reader(fin, delimiter=",", quotechar='"', doublequote=True, escapechar="\\")
    w = csv.writer(fout, delimiter=",", quotechar='"', quoting=csv.QUOTE_MINIMAL)

    header = next(r)
    ncols = len(header)
    body_idx = header.index("Body (HTML)")
    w.writerow(header)

    for row in r:
        # If a row has too many columns, glue the extras back into Body (HTML)
        if len(row) > ncols:
            # keep merging the cell immediately after Body into Body until lengths match
            while len(row) > ncols and body_idx + 1 < len(row):
                row[body_idx] = row[body_idx] + "," + row.pop(body_idx + 1)
        # If a row has too few columns, pad with empty strings (rare)
        if len(row) < ncols:
            row += [""] * (ncols - len(row))
        w.writerow(row)

# Now read the repaired file normally
df = pd.read_csv(cleaned, encoding="utf-8-sig", dtype=str)
print(df.shape)
df.head()
```
